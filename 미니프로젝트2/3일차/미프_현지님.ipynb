```python
train = pd.read_csv("train.csv")
train
```

```python
test = pd.read_csv("test.csv")
test
```

```python
train.isnull().sum()
```

```python
train.describe()
```

```python
train['url_path_len'].fillna(train['url_path_len'].mean(), inplace = True)
train['url_domain_len'].fillna(train['url_domain_len'].mean(), inplace = True)
```

```python
train.describe()
```

```python
train.drop('url_chinese_present', axis=1, inplace = True)
train.drop("html_num_tags('applet')", axis=1, inplace = True)
```

```python
train['label_binary'] = train['label'].copy()
train
```

```python
train = train.replace({'label_binary': {'benign': '0', 'malicious': '1'}})
train
```

```python
train['label_binary'] = pd.to_numeric(train['label_binary'])
```

```python
train.drop('label', axis=1, inplace = True)
```

### 모델링

```python
target = 'label_binary'

# 데이터 분리
x = train.drop(target, axis=1)
y = train.loc[:, target]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=1)
```

### KNN

```python
model = KNeighborsClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
# 예측하기
y_val_pred = model.predict(x_val)
```

```python
print(confusion_matrix(y_val, y_val_pred))
print(classification_report(y_val, y_val_pred))
```

### DecisionTree

```python
model = DecisionTreeClassifier(random_state=1)
```

```python
model.fit(x_train, y_train)
```

```python
# 예측하기
y_val_pred = model.predict(x_val)
```

```python
print(confusion_matrix(y_val, y_val_pred))
print(classification_report(y_val, y_val_pred))
```

### RandomForest

```python
RF = RandomForestClassifier(random_state=1)
```

```python
RF.fit(x_train, y_train)
```

```python
# 예측하기
y_val_pred = RF.predict(x_val)
```

```python
print(confusion_matrix(y_val, y_val_pred))
print(classification_report(y_val, y_val_pred))
```

### XGB

```python
model = XGBClassifier(random_state=1)
```

```python
model.fit(x_train, y_train)
```

```python
# 예측하기
y_val_pred = model.predict(x_val)
```

```python
print(confusion_matrix(y_val, y_val_pred))
print(classification_report(y_val, y_val_pred))
```

```python
from sklearn.model_selection import cross_val_score

# 성능예측
cv_score = cross_val_score(model, x_train, y_train, cv=5)

# 결과
print(cv_score)
print('평균:', cv_score.mean())
```
