## 문제
문자열에 대해 공부하던 기성이는 팰린드롬이라는 개념을 배웠습니다. 팰린드롬이란 앞에서부터 읽어도 뒤에서부터 읽어도 똑같은 문자열입니다. 예를 들어 "MADAM", "RACECAR", "K"는 팰린드롬이고, "MAMA", "AIVLE"은 팰린드롬이 아닙니다.

기성이는 팰린드롬에 엄청난 매력을 느끼고, 팰린드롬의 날을 지정하려고 합니다. 어느 날을 팰린드롬의 날로 정할지 고민하다가, 다음과 같은 날을 고르기로 했습니다.

어떤 날이 A월 B일이라면, A와 B를 붙인 문자열을 C라고 합시다. 단, B가 한자릿수라면 B 앞에 0을 붙여야 합니다. 예를 들어 3월 5일은 C = "305", 4월 12일은 C = "412", 10월 3일은 C = "1003", 11월 11일은 C = "1111"입니다. C가 팰린드롬인 날들 중, 오늘이 아니면서 오늘로부터 가장 먼저 찾아오는 날을 팰린드롬의 날로 정합니다. 오늘이 몇 월 며칠인지는 입력에서 주어집니다.

1월부터 12월까지의 말일은 각각 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31일입니다. 2월 29일은 C = "229"인데 팰린드롬이 아니므로 윤년은 고려하지 않아도 됩니다. 오늘이 몇 월 며칠인지 주어지면, 기성이가 팰린드롬의 날로 골라야 하는 날을 출력하는 프로그램을 작성해 주세요.


### 예제 입력1

5 4

### 예제 출력1

5 5

### 예제 입력2

10 1

### 예제 출력2

11 11

### 입력값 설명

두 정수 A와 B가 공백으로 구분되어 주어집니다.

A월 B일로 읽었을 때, 올바른 날이 되는 입력만 주어집니다.

### 출력값 설명

두 정수 A와 B가 공백으로 구분되어 주어집니다.

A월 B일로 읽었을 때, 올바른 날이 되는 입력만 주어집니다.두 정수 X와 Y를 공백으로 구분하여 출력합니다.

X월 Y일은 기성이가 팰린드롬의 날로 골라야 하는 날이어야 하며, X또는 Y가 한자릿수라도 앞에 0을 붙이지 않습니다.

#### 풀이
- 팰린드롬이란 앞에서부터 읽어도 뒤에서부터 읽어도 똑같은 문자열

```python
A, B = map(int, input().split())
# 날짜 체크
while True:
    B += 1
    if A == 1:
        if B % 10 == 1:
            break
        elif B > 31:
            A = 2
            B = 1
    elif A == 2:
        if B % 10 == 2:
            break
        elif B > 28:
            A = 3
            B = 1
    elif A == 3:
        if B % 10 == 3:
            break
        elif B > 31:
            A = 4
            B = 1
    elif A == 4:
        if B % 10 == 4:
            break
        elif B > 30:
            A = 5
            B = 1
    elif A == 5:
        if B % 10 == 5:
            break
        elif B > 31:
            A = 6
            B = 1
    elif A == 6:
        if B % 10 == 6:
            break
        elif B > 30:
            A = 7
            B = 1
    elif A == 7:
        if B % 10 == 7:
            break
        elif B > 31:
            A = 8
            B = 1
    elif A == 8:
        if B % 10 == 8:
            break
        elif B > 31:
            A = 9
            B = 1
    elif A == 9:
        if B % 10 == 9:
            break
        elif B > 30:
            A = 10
            B = 1
    elif A == 10:
        if B == 1:
            break
        elif B > 31:
            A = 11
            B = 1
    elif A == 11:
        if B == 11:
            break
        elif B > 30:
            A = 12
            B = 1
    elif A == 12:
        if B == 21:
            break
        elif B > 31:
            A = 1
            B = 1

print(A, B)
```

```python
import sys
input = sys.stdin.readline

# 1. 각 월의 말일이 담긴 리스트 생성
days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
# 2. 오늘 날짜 입력받기
month, day = map(int, input().split())
# 3. 10월 미만일 경우
if month < 10 :
    # 3-1. for문
    for potential_day in range(month, days[month], 10) :
        # 3-2. 팰린드롬이 가능한 날짜 확인
        if potential_day > day :
            print(month, potential_day)
            break
    # 3-3. 가능한 날짜가 없다면 다음월의 첫 날 출력
    else :
        if month == 9 : print(month+1, 1)
        else : print(month+1, month+1)
# 10월 이후일 경우
else :
    # 4-1. 해당 월을 뒤집은 문자열이 오늘 날짜보다 크면 출력
    if int(str(month)[::-1]) > day :
        print(month, str(month[::-1]))
    # 4-2. 오늘 날짜가 더 큰 경우 다음 월 출력
    else :
        # 4-3. 12월의 경우 1월 1일 출력
        if month == 12 : print(1, 1)
        else : print(month+1, str(month+1)[::-1])
```
