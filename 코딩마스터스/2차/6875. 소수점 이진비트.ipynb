## 문제
어떤 양의 실수가 주어졌을 때 소수점을 기준으로 실수를 정수부와 소수부 두 부분으로 나눌 수 있습니다. 소수점. 을 기준으로 왼쪽은 정수부가 되고 오른쪽은 소수부가 됩니다. 예를 들어 실수 2.625 가 주어졌다고 합시다. 그러면 이 수의 정수부는 2, 소수부는 0.625입니다.

0보다 크고 1보다 작은 어떤 실수 N의 이진법 표현은 다음과 같은 알고리즘으로 구할 수 있습니다.

1. N에 2를 곱하고, 정수부를 "0."오른쪽에 덧붙여 기록합니다.

2. (1) 결과의 소수부에 2를 곱하고, 정수부를 기록의 오른쪽에 덧붙여 기록합니다.

3. (2) 결과의 소수부에 2를 곱하고, 정수부를 기록의 오른쪽에 덧붙여 기록합니다.

4. (3) 결과의 소수부에 2를 곱하고, 정수부를 기록의 오른쪽에 덧붙여 기록합니다.

... 이와 같은 과정을 소수부가 0이 될 때까지 반복합니다.

예를 들어 0.625의 이진법 표현을 다음과 같이 구할 수 있습니다.

1. 0.625에 2를 곱하면 정수부가 1이고 소수부가 0.25입니다. 현재까지의 기록은 0.1입니다.

2. 0.25에 2를 곱하면 정수부가 0이고 소수부가 0.5입니다. 현재까지의 기록은 0.10입니다.

3. 0.5에 2를 곱하면 정수부가 1이고 소수부가 0.0입니다. 현재까지의 기록은 0.101입니다.

소수부가 0이 되었으므로 더이상 진행하지 않습니다. 0.625의 이진법 표현은 0.101입니다.

한편, 어떤 수의 이진법 표현은 무한히 깁니다. 예를 들어 0.3의 이진법 표현은 0.0100110011001...으로 1001이 무한히 반복됩니다.

0보다 크고 1보다 작은 어떤 실수 N의 이진법 표현을 출력하는 프로그램을 작성해 주세요. 만약 출력이 소수점 11자리를 넘는다면 소수점 11자리 이상은 버리고 소수점 10자리까지만 출력하는 프로그램을 작성해 주세요.


### 예제 입력1

0.625

### 예제 출력1

0.101

### 예제 입력2

0.3

### 예제 출력2

0.0100110011

### 입력값 설명

첫째 줄에 실수 N이 주어집니다.(0 < N < 1)

### 출력값 설명

실수 N의 이진법 표현을 출력하세요. 만약 출력이 소수점 11자리를 넘는다면 소수점 11자리 이상은 버리고 소수점 10자리까지만 출력하세요.

#### 풀이
- 실수N의 이진법 표현
- N * 2 -> 정수부분만 추출하여 append
- 소수점 10자리까지만 출력

```python
N = float(input())
answer = 0
for i in range(1,11):
    if N == 0:
        break
    N *= 2
    k = int(N)
    N -= k
    answer += (0.1**i) * k
print(answer)
# 소수점 0.0000으로 된건 제대로 출력이 안된다.
```
